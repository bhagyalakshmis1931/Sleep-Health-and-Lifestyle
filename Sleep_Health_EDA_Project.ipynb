### 11. Condition Based Filtering & Visualization


##### Listing occupation with high stress level (stress level greater than 8)


```python
high_stress=df[df['Stress_Level']>=8]
high_stress.groupby('Occupation')['Occupation'].count()
```

##### Number of people have sleep disorders

```python
df_sleepdisorder= df.query("Sleep_Disorder == 'Sleep Apnea'| Sleep_Disorder == 'Insomnia' ") 
df_sleepdisorder.groupby('Sleep_Disorder')['Sleep_Disorder'].count()
```

##### Here we categorizing the column 'Quality_of_Sleep' into three ('Poor', 'Fair', 'Good'), then plotting bar chart with the frequency count of each category

```python
bins = [0, 5, 7, 10]  
labels = ['Poor', 'Fair', 'Good']
df['Quality of Sleep Category'] = pd.cut(df['Quality_of_Sleep'], bins=bins, labels=labels, include_lowest=True)
df.groupby('Quality of Sleep Category')['Quality of Sleep Category'].count().plot(kind='bar',color=['#9AD0C2','#2D9596', '#265073'])
plt.xlabel('Quality of sleep')
plt.ylabel('Frequency')
plt.title('Quality of Sleep Categories')
plt.xticks(rotation=0)
plt.show()
```

##### Grouping the dataset by  Sleep_Disorder

```python
sleep_stress_df = df.groupby('Sleep_Disorder').agg(stress=('Stress_Level','mean'),sleep_quality=('Quality_of_Sleep','mean'),sleep_duration=('Sleep_Duration','mean'))
sleep_stress_df.T.plot(kind='bar',color=['#45474B','#748E63','#F4CE14'],stacked=False)
plt.xticks(rotation=0)
plt.legend()
plt.show()
```

##### Representing different types of sleep disorders using pie chart

```python
df.groupby('Sleep_Disorder')['Sleep_Disorder'].agg('count').plot(kind='pie',colors=['#FFB400','#2D4059','#00917C'],autopct='%.2f%%')
plt.ylabel(" ")
plt.legend(loc='lower right',fontsize=8)
plt.show()
```

##### Comparing stress level and quality of sleep in adult and middle aged persons 

```python
bins = [25,44,60]  
labels = ['Adult', 'Middle Aged']
df['Age_Category'] = pd.cut(df['Age'], bins=bins, labels=labels, include_lowest=True)
df.groupby('Age_Category')[['Stress_Level','Quality_of_Sleep']].mean().plot(kind='barh',stacked=True,color=['#FB8B24','#7E2553'])
plt.ylabel("Age Category")
plt.xlabel("Frequency")
plt.legend()
plt.show()
```

##### Plotting the mean stress level among different job roles using vertical bar chart

```python
df.groupby('Occupation')['Stress_Level'].mean().plot(kind='barh',color=['#6D2932','#F35F5F','#DE5B7B','#FFD3B6','#FFF87D','#69C181','#B36458',
                                                                        '#B34180','#BEE5D3','#6E85B7','#17594A'])
plt.x()
plt.legend()
plt.show()
```

##### Visualizing mean stress levels among males and females using histogram

```python

pd.pivot_table(df,values='Stress_Level',columns='Gender',aggfunc='mean').plot(kind='hist',color=['#FFB996','#9BB8CD'])
plt.xlabel("Gender", fontweight='bold')
plt.ylabel(" ")
plt.legend()
plt.show()
```

##### Calculating the count of different types of sleep disorders with respect to the BMI Category and plotting it by multiple barchart

```python
BMI_based_sleepdisorder=df.groupby('BMI_Category')['Sleep_Disorder'].value_counts()
BMI_based_sleepdisorder
```

```python
categories = ['Normal', 'Normal Weight', 'Overweight','Obese']
NONE = [183, 17, 19,0]
INSOMNIA =  [7, 2, 64, 4]
SLEEP_APNEA =  [5, 2, 65, 6]
X_axis = np.arange(len(categories))
plt.barh(X_axis +0.1, NONE, height=0.2,  label = 'NONE',color='#B4B4B3')
plt.barh(X_axis +0.2, INSOMNIA, height=0.2,label = 'INSOMNIA',color='#26577C')
plt.barh(X_axis +0.3, SLEEP_APNEA, height=0.2,  label = 'SLEEP_APNEA',color='#80BCBD')
plt.yticks(X_axis, categories)
plt.xlabel("categories", fontsize=11, fontweight='bold')
plt.ylabel("No.of Persons", fontsize=11, fontweight='bold')
plt.legend()
plt.show()
```
